```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

Task 1: Choose a greyscale image

```python
img = data.moon().astype(np.float32) / 255.0

# Compute DFT and shift
F = np.fft.fft2(img)
F_shift = np.fft.fftshift(F)
# Filter functions definitions
def ideal_lowpass(shape, D0):
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    U, V = np.ogrid[:rows, :cols]
    D = np.sqrt((U - crow)**2 + (V - ccol)**2)
    return (D <= D0).astype(np.float32)

def butterworth_lowpass(shape, D0, n=2):
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    U, V = np.ogrid[:rows, :cols]
    D = np.sqrt((U - crow)**2 + (V - ccol)**2)
    return 1 / (1 + (D / (D0 + 1e-12))**(2*n))

def gaussian_lowpass(shape, D0):
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    U, V = np.ogrid[:rows, :cols]
    D2 = (U - crow)**2 + (V - ccol)**2
    return np.exp(-D2 / (2 * (D0**2)))
```

Task 2: Apply low pass filters

```python
D0 = 35

# Apply filters
H_ideal = ideal_lowpass(img.shape, D0=D0)
H_butt = butterworth_lowpass(img.shape, D0=D0, n=2)
H_gauss = gaussian_lowpass(img.shape, D0=D0)

out_ideal = np.real(np.fft.ifft2(np.fft.ifftshift(F_shift * H_ideal)))
out_butt = np.real(np.fft.ifft2(np.fft.ifftshift(F_shift * H_butt)))
out_gauss = np.real(np.fft.ifft2(np.fft.ifftshift(F_shift * H_gauss)))

# Show results
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(out_ideal, cmap="gray")
plt.title("Ideal LPF")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(out_butt, cmap="gray")
plt.title("Butterworth LPF")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(out_gauss, cmap="gray")
plt.title("Gaussian LPF")
plt.axis("off")
plt.show()
```

Task 3: Apply a high pass filter and sharpen

```python
# Construct high pass from gaussian
H_hp = 1 - H_gauss
out_hp = np.real(np.fft.ifft2(np.fft.ifftshift(F_shift * H_hp)))

# Sharpening
sharp = np.clip(img + 0.7 * out_hp, 0, 1)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(out_hp, cmap="gray")
plt.title("High pass result")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(sharp, cmap="gray")
plt.title("Sharpened image")
plt.axis("off")
plt.show()
```

Task 4: Create periodic noise and remove it

```python
rows, cols = img.shape
x = np.arange(cols)

# Add periodic noise
noise = 0.10 * np.sin(2 * np.pi * x / 20)
noise2d = np.tile(noise, (rows, 1)).astype(np.float32)
img_p = np.clip(img + noise2d, 0, 1)

# DFT for noisy image
Fp = np.fft.fftshift(np.fft.fft2(img_p))

# Notch filter mask
Hn = np.ones_like(img_p, dtype=np.float32)
crow, ccol = rows // 2, cols // 2
notches = [(crow, ccol+26), (crow, ccol-26)] # Adjusted for our noise
rad = 4
U, V = np.ogrid[:rows, :cols]

for (u0, v0) in notches:
    D = np.sqrt((U - u0)**2 + (V - v0)**2)
    Hn[D <= rad] = 0

out_notch = np.real(np.fft.ifft2(np.fft.ifftshift(Fp * Hn)))

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_p, cmap="gray")
plt.title("Corrupted")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(Hn, cmap="gray")
plt.title("Notch reject mask")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(out_notch, cmap="gray")
plt.title("After notch filtering")
plt.axis("off")
plt.show()
```

**Reflection questions**

1- Why is frequency domain filtering often more intuitive for periodic noise?

Because periodic noise shows up as clear bright dots in the frequency domain. We can easily block these dots using a notch mask without destroying the image.

2- Why does sharp cutoff in the spectrum lead to spatial ringing?

A sharp cutoff drops frequencies to zero suddenly. When we apply the inverse FFT, this sudden drop creates ripples around the image edges, causing the ringing effect.

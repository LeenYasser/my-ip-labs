Task 1: Setup & Clean Image

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, util
from scipy.ndimage import uniform_filter, median_filter, generic_filter

# Load image
img = data.moon().astype(np.float32) / 255.0

# MSE
def mse(a, b):
    a = np.float32(a)
    b = np.float32(b)
    return float(np.mean((a - b)**2))

# Geo Mean
def geometric_mean_filter(im, size=3, eps=1e-6):
    im2 = np.clip(im, eps, 1.0)
    log_im = np.log(im2)
    m = uniform_filter(log_im, size=size)
    return np.exp(m)

# Contraharmonic
def contraharmonic(im, size=3, Q=1.5, eps=1e-6):
    def f(window):
        w = np.array(window, dtype=np.float32)
        num = np.sum((w + eps) ** (Q + 1))
        den = np.sum((w + eps) ** Q)
        return num / (den + eps)
    return generic_filter(im, f, size=size)
```

Task 2: Gaussian Noise

```python
# Gaussian
rng = np.random.default_rng(0)
sigma = 0.08
noise = rng.normal(0, sigma, img.shape).astype(np.float32)
g = np.clip(img + noise, 0, 1)

mean3 = uniform_filter(g, size=3)
geo3 = geometric_mean_filter(g, size=3)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(g, cmap="gray"), plt.title("Gaussian"), plt.axis("off")
plt.subplot(1, 3, 2), plt.imshow(mean3, cmap="gray"), plt.title("Mean"), plt.axis("off")
plt.subplot(1, 3, 3), plt.imshow(geo3, cmap="gray"), plt.title("Geo Mean"), plt.axis("off")
plt.show()
```

Task 3: Salt and Pepper Noise

```python
# S&P
sp = util.random_noise(img, mode="s&p", amount=0.08).astype(np.float32)

med3 = median_filter(sp, size=3)
contra = contraharmonic(sp, size=3, Q=1.5)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(sp, cmap="gray"), plt.title("S&P"), plt.axis("off")
plt.subplot(1, 3, 2), plt.imshow(med3, cmap="gray"), plt.title("Median"), plt.axis("off")
plt.subplot(1, 3, 3), plt.imshow(contra, cmap="gray"), plt.title("Contra"), plt.axis("off")
plt.show()
```

Task 4: Evaluation & Comparison

```python
# Print MSE
print("Gaussian MSE:", mse(img, g))
print("Mean MSE:", mse(img, mean3))
print("Geo Mean MSE:", mse(img, geo3))
print("S&P MSE:", mse(img, sp))
print("Median MSE:", mse(img, med3))
print("Contra MSE:", mse(img, contra))

```

**Reflection Questions**

1- Why is restoration more objective than enhancement?

Restoration uses math to reverse a specific noise model. Enhancement just changes the image to look "nice" to human eyes, which is subjective.

2- Why must noise models be identified before restoration?

Because different noises need different filters. For example, a Mean filter fixes Gaussian noise, but a Median filter is needed for Salt & Pepper noise.

3- Why is there no single best denoising filter?

Because there is always a trade-off. A filter that removes noise very well usually blurs the edges and loses small details.
